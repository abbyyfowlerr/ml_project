```python
from radiomics import featureextractor
import SimpleITK as sitk
from PIL import Image
import numpy as np

# Define the paths (assuming they are correct now)
mask_path = "/Users/abby/.cache/kagglehub/datasets/aryashah2k/breast-ultrasound-images-dataset/versions/1/Dataset_BUSI_with_GT/benign/benign (1)_mask.png"
image_path = "/Users/abby/.cache/kagglehub/datasets/aryashah2k/breast-ultrasound-images-dataset/versions/1/Dataset_BUSI_with_GT/benign/benign (1).png"

mask = sitk.ReadImage(mask_path)
mask = sitk.Cast(mask, sitk.sitkUInt8)

image = Image.open(image_path)
grayscale_image = image.convert("L")
np_img = np.array(grayscale_image)
sitk_image = sitk.GetImageFromArray(np_img)

params = "settings.yml"

extractor = featureextractor.RadiomicsFeatureExtractor(params)
result = extractor.execute(sitk_image, mask)
print(result)
```
